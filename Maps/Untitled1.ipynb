```python
import numpy as np
from scipy import interpolate
from bokeh.models import Circle, ColumnDataSource, CustomJS
from bokeh.plotting import figure, save
from bokeh.io import output_file, show, output_notebook, output_file
from bokeh.events import Event, LODEnd
import geopandas as gpd
import requests
```

```python
x = np.linspace(0,10,10)
y = np.linspace(0,10,10)
xgrid, ygrid = np.meshgrid(x,y)
vals = 5*np.random.sample(100)
```

```python
interp = interpolate.RectBivariateSpline(x, y, vals.reshape(10,10))
```

```python
# Create ColumnDataSource from coordinates, sizes and predicted means
source = ColumnDataSource(dict(x=xgrid.reshape(100), 
                               y=ygrid.reshape(100), 
                               vals=vals))

# Plot figure and add tiles 
map = figure(title=None, x_range=(0,10), y_range=(0,10), tools = "pan,wheel_zoom,box_zoom,reset,hover")

# Create and plot glyphs for predictions
glyph = Circle(x="x", y="y", radius= 0.1, line_color="black", fill_color="black", fill_alpha=0.8, line_width=1, 
                line_alpha=1)
map.add_glyph(source, glyph)

callback = CustomJS(args=dict(source=source, xr=map.x_range, yr=map.y_range), code="""
    
    var xstart = xr.start;
    var xend = xr.end;
    var ystart = yr.start;
    var yend = yr.end;
    var gridsize = 10;
        
    source.change.emit();          
    """)

map.js_on_event(LODEnd, callback)

output_notebook()
output_file()
show(map)
# Save the plot by passing the plot -object and output path
#save(obj=scaled_map, filename=outfp)
```
